<h2>Q3: Part 1: Pairwise Plotting of Features</h2>

```python
# Import the packages that are required for this section
# Note: These packages are imported further here instead of at the beginning of the notebook
# as seaborn does not display outliers on the plots, which are needed for Question 2
import matplotlib.patches as mpatches
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns
```

```python
# Correlation matrix using code found on https://stanford.edu/~mwaskom/software/seaborn/examples/many_pairwise_correlations.html
CleanDataFile = "quality_issues_addressed_amazon_information.csv"
df_clean = pd.read_csv(CleanDataFile)

sns.set(style="white")

# Select columns containing continuous data
continuous_columns = df_clean[['ListingPrice','SellerFeedbackRating','ShippingPrice','ShippingTime_minHours','SellerFeedbackCount', 'ShippingTime_maxHours']].columns

# Calculate correlation of all pairs of continuous features
corr = df_clean[continuous_columns].corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom colormap - blue and red
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=1, vmin=-1,
            square=True, xticklabels=True, yticklabels=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.yticks(rotation = 0)
plt.xticks(rotation = 45)
```

```python
# Find unique values in Fulfilled by Amazon
countries = pd.unique(df_clean.IsFulfilledByAmazon.ravel())

# Insert a new column called 'percent' and fill it with 0s
df_clean['percent'] = 0

# Iterate through the unique values in Fulfilled by Amazon and for each value count the amount of Winners
# Find the indexes of each row and for each of these row insert count * 100 in the percent column
for c in countries:
    count = 1 / df_clean[df_clean.IsFulfilledByAmazon == c].count()['IsWinner']
    index_list = df_clean[df_clean['IsFulfilledByAmazon'] == c].index.tolist()
    for i in index_list:
        df_clean.loc[i, 'percent'] = count * 100
        
# Group dataframe by Fulfilled by Amazon and IsWinner and sum percent
group = df_clean[['percent','IsFulfilledByAmazon','IsWinner']].groupby(['IsFulfilledByAmazon','IsWinner']).sum()

# Plot values of group in a stacked bar chart
my_plot = group.unstack().plot(kind='bar', stacked=True, title="IsWinner is or is not a IsFulfilledByAmazon", figsize=(15,7))

# Define label colours and text and add to the plot
red_patch = mpatches.Patch(color='green', label='1')
blue_patch = mpatches.Patch(color='blue', label='0')
my_plot.legend(handles=[red_patch, blue_patch], frameon = True)

# Define x and y labels and min and max values for the y axis
my_plot.set_xlabel("IsFulfilledByAmazon")
my_plot.set_ylabel("% IsWinner")
my_plot.set_ylim([0,100])
```
